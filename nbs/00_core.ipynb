```python
#|export
def add(a, b):
    return a + b

#|export
def subtract(a, b):
    return a - b

#|export
def multiply(a, b):
    return a * b

#|export
def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b

#|export
def calculator_logic(a, b, op):
    if op == '+':
        return add(a, b)
    elif op == '-':
        return subtract(a, b)
    elif op == '*':
        return multiply(a, b)
    elif op == '/':
        return divide(a, b)
    else:
        return "Invalid operator"
```
